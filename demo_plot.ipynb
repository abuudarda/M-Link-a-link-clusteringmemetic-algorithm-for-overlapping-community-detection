```python
import networkx as nx
import community
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Create a sample graph
G = nx.karate_club_graph()

# Use the Louvain algorithm to detect communities
partition = community.best_partition(G)

# Create a list of colors for the nodes based on their communities
colors = [partition[node] for node in G.nodes()]

# Create a layout for the nodes
pos = nx.spring_layout(G)

# Create a figure and axis
fig, ax = plt.subplots()

# Draw the graph with node colors based on communities
nx.draw(G, pos, node_color=colors, with_labels=True, cmap=plt.cm.get_cmap("viridis"))

# Create background figures for each community
for community_id in set(partition.values()):
    nodes_in_community = [node for node in G.nodes() if partition[node] == community_id]
    x, y = zip(*[pos[node] for node in nodes_in_community])
    ax.add_patch(patches.Polygon(xy=list(zip(x, y)), closed=True, fill=True, color='lightgray'))

plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a graph
G = nx.Graph()

# Add nodes to the graph
nodes = ["A", "B", "C", "D", "E", "F"]
G.add_nodes_from(nodes)

# Add edges to the graph
edges = [("A", "B"), ("B", "C"), ("C", "D"), ("D", "E"), ("E", "F"), ("A", "F")]
G.add_edges_from(edges)

# Create subgraphs
subgraph1 = ["A", "B", "C"]
subgraph2 = ["D", "E", "F"]

# Create a list of subgraphs
subgraphs = [subgraph1, subgraph2]

# Create the layout (you can customize it as needed)
pos = nx.spring_layout(G)

# Create a figure and axis for the visualization
fig, ax = plt.subplots()

# Draw the graph
nx.draw(G, pos=pos, with_labels=True, ax=ax)

# Draw background shapes for subgraphs
for subgraph in subgraphs:
    min_x = min([pos[node][0] for node in subgraph])
    max_x = max([pos[node][0] for node in subgraph])
    min_y = min([pos[node][1] for node in subgraph])
    max_y = max([pos[node][1] for node in subgraph])

    ax.add_patch(plt.Rectangle((min_x, min_y),
                              max_x - min_x,
                              max_y - min_y,
                              fill=False, linewidth=2, edgecolor='blue'))

# Show the plot
plt.show()
```
